```python
# from notebook_setup import PROJECT_ROOT
import notebook_setup
import gzip
import json
import pandas as pd
import random
import os
import re

from utils.path_config import get_data_path


# Customize these for each category

# def get_data_path(rel_path):
#     return PROJECT_ROOT / rel_path

# # Customize these for each category
# Change for Beauty/Fashion, etc.
input_path = get_data_path("Electronics_23.json.gz") #using utils.path_config.get_data_path
# input_path = get_data_path("data/Electronics_23.json.gz")

# Where to save your sample   
output_path = get_data_path("Clothing_Shoes_and_Jewelry_subset.csv")  #using utils.path_config.get_data_path
# output_path = get_data_path("data/Clothing_Shoes_and_Jewelry_subset.csv")

sample_size = 25000                         # Sample rows per category

# First, count total lines
with gzip.open(input_path, "rt", encoding="utf-8") as f:
    total_lines = sum(1 for _ in f)
print(f"Total lines in file: {total_lines}")

# Pick random unique indices
random_indices = set(random.sample(range(total_lines), sample_size))

rows = []
with gzip.open(input_path, "rt", encoding="utf-8") as f:
    for i, line in enumerate(f):
        if i in random_indices:
            try:
                rows.append(json.loads(line))
            except json.JSONDecodeError:
                continue
        if len(rows) >= sample_size:
            break

df = pd.DataFrame(rows)
print(f"Randomly sampled {len(df)} rows. Columns: {df.columns.tolist()}")
df.to_csv(output_path, index=False)
print(f"Sample saved to {output_path}")
```
